```python
import requests
import numpy as np
from bs4 import BeautifulSoup
import pandas as pd
```

```python
items_links_women = []
```

```python
url = "https://www.armedangels.com/fr-fr/femme/meilleures-ventes"
head = {"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/109.0"}
r = requests.get(url, headers  = head)
print(r.status_code)


items_links_women = []

soup = BeautifulSoup(r.text, "html.parser")
links = soup.find_all("a",  {"class":"product-info-link js-product-info-link gtm-product-link"})

for link in links:
    href = link.get('href')
    items_links_women.append(href)




url = "https://www.armedangels.com/fr-fr/homme/meilleures-ventes"
head = {"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/109.0"}
r = requests.get(url, headers  = head)
print(r.status_code)

soup = BeautifulSoup(r.text, "html.parser")
links = soup.find_all("a",  {"class" : "product-info-link js-product-info-link gtm-product-link"})



items_links_men = []
for link in links:
    href = link.get('href')
    items_links_men.append(href)

list_items = items_links_women + items_links_men

list_items = list(set(list_items))

cloths = []
color = []
country = []
factory = []
price = []
gender = []
product_id_clothes = []

material = []
percentage = []
label1, label2, label3 = [], [], []

labels_list = [label1, label2, label3]
product_id_composition = []



for item_url in list_items:
    
    # Calculez la valeur absolue du hachage de l'URL
    url_hash = abs(hash(url))

    # Utilisez le modulo pour obtenir un nombre entier entre 1000 et 9999
    product_id = 1000 + (url_hash % 9000)
    
    product_id_clothes.append(product_id)
    
    
    print(list_items.index(item_url))
    url = item_url
    print(url)
    head = {"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/109.0"}
    r = requests.get(url, headers  = head, timeout = None)
    
    
    # Parse the HTML content
    soup = BeautifulSoup(r.text, "html.parser")
    
    #Scrapping cloth
    cloths.append(soup.find("title").text.lstrip().split(" ")[0])
    print(soup.find("title").text.lstrip().split(" ")[0])
    
    #Scrapping Price
    price.append(soup.find("p",{"class": "product-price"}).text.split("0")[0])
    print(soup.find("p",{"class": "product-price"}).text.split("0")[0])
    
    if url in items_links_women:    
        gender.append("Femme")
        print("Femme")
    if url in items_links_men:
        gender.append("Homme")
        print("Homme")
        
    
    #Scrapping color
    color.append(soup.find("p", {"class" : "body-text swatch-group__headline"}).text)
    print(soup.find("p", {"class" : "body-text swatch-group__headline"}).text)
    
    pays = np.nan
    producteur = np.nan
    
    #Scrapping country and company
    ul_tag = soup.find('ul', {'class': 'body-text mb-0 list-styled'})
    li_items = ul_tag.find_all('li')
    for li in li_items:
        if 'Lieu de production:' in li.text:
            pays = li.span.text
        
        if 'Producteur:' in li.text:
            producteur = li.span.text
            
    print(pays)
    print(producteur)
    country.append(pays)
    factory.append(producteur)
    
    
    
    # Find the dl element containing the composition information
    composition = soup.find("div", {"class" : "material-text"}).text.split(", ")
    print(composition)
    
    for material_str in composition:
        
        material_percentage = material_str.split(" ")[0]
        print(material_percentage)
        material_str = " ".join(material_str.split(" ")[1:]).lstrip()
        print(material_str)
        percentage_float = float(material_percentage.strip("%"))
        print(percentage_float)
        
        # Append the details to the lists
        product_id_composition.append(product_id)
        material.append(material_str)
        percentage.append(percentage_float)
        
        
  # Scrapping certification
    certifications = soup.find_all("div", {'class': "certification"})



    for i, certification in enumerate(certifications):
        label_p = certification.find("p")
        text = label_p.decode_contents()
        first_line = text.split("<br", 1)[0].strip()

        if i < len(labels_list):
            labels_list[i].append(first_line)
        else:
            print(f"Index {i} is out of range for labels_list")

    for i in range(len(certifications), len(labels_list)):
        labels_list[i].append(np.nan)

    for i, labels in enumerate(labels_list):
        print(f"{i}: {labels}")


   

    
```

```python
# Create the dataframe
df_armed_composition = pd.DataFrame({
"product_id": product_id_composition,
"material": material,
"percentage": percentage
})

df_armed_cloth = pd.DataFrame({
    "product_id" : product_id_clothes,
    "gender" : gender,
    "clothes" : cloths,
    "color" : color,
    "price" : price,
    "label1" : label1,
    "label2" : label2,
    "label3" : label3
    
})

df_armed_supplier = pd.DataFrame({
    "product_id" : product_id_clothes,
    "company" : factory,
    "country" : country
})
```
