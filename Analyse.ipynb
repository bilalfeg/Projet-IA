```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport 
```

```python
df_tendance_composition = pd.read_csv("data_h&m_tendance_compo.csv")
df_tendance_clothes = pd.read_csv("data_h&m_tendnace_clothes.csv")
higgs_score = pd.read_csv("higgs_score_materials.csv")
```
